```python
# Function to extract features from the model column
def extract_model_features(model):
    """
    This function extracts features from the model column creating new features 
    and enriching the dataset.
    """
    parts = model.split()
    base_model = parts[0] if parts else model
    model_variant = parts[1] if len(parts) > 1 else ''
    additional_features = ''.join(parts[2:]) if len(parts) > 2 else ''
    return base_model, model_variant, additional_features
```

```python
# Applying extraction function to 'model' column
train_tmp[['base_model', 'model_variant', 'additional_features']] = train_tmp['model'].apply(lambda x: pd.Series(extract_model_features(x)))
train_tmp
```

```python
# Function to extract features from the engine column 
def extract_engine_features(engine_str):
    # Extract displacement
    displacement = re.search(r"(\d+\.\d+)L", engine_str)
    engine_displacement = float(displacement.group(1)) if displacement else None

    # Extract horsepower
    horsepower = re.search(r"(\d+\.\d+)HP", engine_str)
    engine_horsepower = float(horsepower.group(1)) if horsepower else None

    # Extract number of cylinders
    num_cylinders = re.search(r"\d+\s?Cylinder", engine_str)
    num_cylinders = int(num_cylinders.group().split()[0]) if num_cylinders else None

    # Extract cylinder configuration
    cylinder_config = "V" if "V" in engine_str else "Straight" if "Straight" in engine_str else None

    # # Extract fuel type
    # fuel_type = re.search(r"(Gasoline|Diesel|Electric|Flex Fuel)", engine_str)
    # fuel_type = fuel_type.group(1) if fuel_type else None

    return engine_displacement, engine_horsepower, num_cylinders, cylinder_config
```

```python
# Creating new columns for extracted features 
train_tmp[['engine_displacement', 'engine_horsepower', 'num_cylinders', 'cylinder_config']] = train_tmp['engine'].apply(extract_engine_features).apply(pd.Series)
train_tmp
```

```python
# Function to extract features from transmission 
def extract_transmission_features(transmission_str):
    if 'A/T' in transmission_str or 'Automatic' in transmission_str or 'Auto' in transmission_str or 'automatic' in transmission_str or 'auto' in transmission_str:
        transmission_type = 'A/T'
        num_gears = re.search(r'(\d+)-Speed', transmission_str)
        num_gears = int(num_gears.group(1)) if num_gears else None
        transmission_features = transmission_str.replace(str(num_gears) + '-Speed A/T', '').strip()
    elif 'M/T' in transmission_str or 'Manual' in transmission_str or 'manual'in transmission_str:
        transmission_type = 'M/T'
        num_gears = re.search(r'(\d+)-Speed', transmission_str)
        num_gears = int(num_gears.group(1)) if num_gears else None
        transmission_features = transmission_str.replace(str(num_gears) + '-Speed M/T', '').strip()
    elif 'Dual' in transmission_str or 'dual' in transmission_str:
        transmission_type = 'Dual'
        num_gears = None
        transmission_features = transmission_str.strip()
    else:
        transmission_type = 'Unknown'
        num_gears = None
        transmission_features = ''
    return transmission_type, num_gears, transmission_features
```

```python
train_tmp[['transmission_type', 'num_gears', 'transmission_features']] = train_tmp['transmission'].apply(extract_transmission_features).apply(pd.Series)
```

### Filling missing values


```python
train_tmp.isna().sum()
```

```python
# Check for which numeric columns have null values 
for label, content in train_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            print(label)
```

```python
# Fill numeric rows with median 
for label, content in train_tmp.items():
    if pd.api.types.is_numeric_dtype(content):
        # Add a binanry column which tells us if data was missing or not
        train_tmp[label + "_is_missing"] = pd.isnull(content)
        # Fill missing numeric values with median 
        train_tmp[label] = content.fillna(content.median())
```

```python
# Turn categorical variables into numerical data using 
for label, content in train_tmp.items():
    if not pd.api.types.is_numeric_dtype(content):
        # Add a binary column to indicate wether sample had missing value or not
        train_tmp[label + "_is_missing"] = pd.isnull(content)
        # Turn categories into numbers and add +1
        train_tmp[label] = pd.Categorical(content).codes + 1
```
